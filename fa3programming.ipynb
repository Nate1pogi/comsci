## Program No. 1 (fa3prog1): Student Information Display

```python
# Program No. 1 (fa3prog1): Student Information Display

# a. Asks for the student’s name
student_name = input("Enter student's name: ")

# b. Asks for the student’s age
student_age = input("Enter student's age: ")

# c. Asks for the student’s course
student_course = input("Enter student's course: ")

# d. Displays all the entered information in a formatted output
print("\n--- Student Information ---")
print(f"Name: {student_name}")
print(f"Age: {student_age}")
print(f"Course: {student_course}")
print("-------------------------")
```

## Program No. 2 (fa3prog2): Pass or Fail Checker

```python
# Program No. 2 (fa3prog2): Pass or Fail Checker

# a. Accepts a student’s exam score
try:
    score = float(input("Enter student's exam score: "))

    # b. Displays “Passed” if the score is 75 or higher
    if score >= 75:
        print("Passed")
    # c. Displays “Failed” if the score is below 75
    else:
        print("Failed")
except ValueError:
    print("Invalid input. Please enter a numeric score.")
```

## Program No. 3 (fa3prog3): Grading System

```python
# Program No. 3 (fa3prog3): Grading System

# Input the student’s name and score
student_name_grade = input("Enter student's name: ")
try:
    score_grade = float(input(f"Enter {student_name_grade}'s score: "))

    # Determine the corresponding letter grade
    if 90 <= score_grade <= 100:
        letter_grade = 'A'
    elif 80 <= score_grade < 90:
        letter_grade = 'B'
    elif 70 <= score_grade < 80:
        letter_grade = 'C'
    else:
        letter_grade = 'F'

    # Display the result
    print(f"Student: {student_name_grade}")
    print(f"Score: {score_grade}")
    print(f"Grade: {letter_grade}")
except ValueError:
    print("Invalid input. Please enter a numeric score.")
```
