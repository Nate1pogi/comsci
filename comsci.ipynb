```python
def get_float_input(prompt):
    while True:
        try:
            value = float(input(prompt))
            return value
        except ValueError:
            print("Invalid input. Please enter a valid number.")

num1 = get_float_input("Enter the first number: ")
num2 = get_float_input("Enter the second number: ")
num3 = get_float_input("Enter the third number: ")

biggest_number = max(num1, num2, num3)
```

```python
print(f"The biggest number is: {biggest_number}")
```
